```python
import numpy as np
from numpy import ndarray

students:ndarray = np.array([[79, 93, 69, 96, 92],
                            [66, 76, 64, 85, 87]])

subjects:list[str] = ['chinese', 'english', 'math', 'history', 'discover']
subjects
['chinese', 'english', 'math', 'history', 'discover']
import matplotlib.pyplot as plt

figure = plt.figure()
axes = figure.add_subplot()
axes.plot(subjects, students[0],label="stu1")
axes.plot(subjects,students[1],label="stu2")
axes.legend()
axes.set_title("stu1 vs stu2")
axes.set_xlabel("subject")
axes.set_ylabel('scores')
plt.show()
plt.close()
```

導入和數據設置：
import numpy as np 和 from numpy import ndarray：導入 NumPy 庫和 ndarray 類。
students: ndarray = np.array([[79, 93, 69, 96, 92], [66, 76, 64, 85, 87]])：創建一個二維 NumPy 陣列，包含兩行（代表兩位學生）和五列（代表五個科目的成績）。
students[0]：第一位學生的成績：[79, 93, 69, 96, 92]。
students[1]：第二位學生的成績：[66, 76, 64, 85, 87]。
subjects: list[str] = ['chinese', 'english', 'math', 'history', 'discover']：定義科目名稱列表，對應 students 陣列的列。
subjects 這一行在交互式環境中會輸出 ['chinese', 'english', 'math', 'history', 'discover']，但在此腳本中無實際影響，因為未賦值或打印。
使用 Matplotlib 繪圖：
import matplotlib.pyplot as plt：導入 Matplotlib 的繪圖模組。
figure = plt.figure()：創建一個新的圖表。
axes = figure.add_subplot()：添加一個子圖（預設為 1x1 網格，等同於 add_subplot(111)）。
axes.plot(subjects, students[0], label="stu1")：繪製第一位學生 (stu1) 的成績折線圖，數據為 [79, 93, 69, 96, 92]，對應科目 ['chinese', 'english', 'math', 'history', 'discover']，標籤為“stu1”。
axes.plot(subjects, students[1], label="stu2")：繪製第二位學生 (stu2) 的成績折線圖，數據為 [66, 76, 64, 85, 87]，標籤為“stu2”。
axes.legend()：顯示圖例，區分“stu1”和“stu2”兩條折線。
axes.set_title("stu1 vs stu2")：設置圖表標題為“stu1 vs stu2”。
axes.set_xlabel("subject")：設置 x 軸標籤為“subject”。
axes.set_ylabel("scores")：設置 y 軸標籤為“scores”。
plt.show()：顯示圖表。
plt.close()：關閉圖表以釋放內存。

備註
由於科目是類別數據（非連續），折線圖假設數據點之間有連續關係。如果需要更適合類別數據的圖表（例如柱狀圖），我可以提供 Chart.js 的柱狀圖程式碼。

```python
import numpy as np
from numpy import ndarray
students=np.array([[96,78,88],
                     [88,80,90]])
subject=['Chinese','English','History']
import matplotlib.pyplot as plt
figure=plt.figure()
axes=figure.add_subplot()
axes.plot(subject,students[0],label="stu1")
axes.plot(subject,students[1],label="stu2")
axes.legend()
axes.set_xlabel("subject")
axes.set_ylabel("score")
plt.show()
plt.close()
```
